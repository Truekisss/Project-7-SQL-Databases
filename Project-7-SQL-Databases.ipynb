### Loading Data from DataFrames into an SQLite Database

```python
import pandas as pd
import json
import sqlite3 as sq3
```

```python
# Opening json file and store the data on the variable data
with open("some_movies.json") as f:
    data = json.load(f)
```

```python
# 80 movies in that data
data
```

```python
# Creating pandas dataframe and flatten it with pd.json_normalize
df = pd.json_normalize(data, sep = "_")
```

```python
df
```

```python
# Creating dataframe with the most important features like id, title etc,
    # and saving this dataframe in the variable movies.
movies = df[["id", "title", "revenue", "budget", "belongs_to_collection_name", "release_date"]].copy()
movies
```

```python
# Futher inspection of movies
movies.info()
```

```python
# Datatype in the real estate column should be a datetime
movies.release_date = pd.to_datetime(df.release_date)
```

```python
# Revenue and budget in terms of U.S. dollars
movies.revenue = df.revenue/100000
movies.budget = df.budget/100000
```

```python
# Creating 2nd dataframe with votes and ratings
votes = df[["id", "vote_count", "vote_average"]].copy()
votes
```

```python
# There is nested data in genres column, thus converting the genre column in to one
    # separated dataframe with pd.json_normalize
# Also linking the new data frame with main table
genres = pd.json_normalize(data = data, record_path = "genres", meta = "id", record_prefix = "genre_")
genres
```

```python
# prod is a nested data, thus creating separate dataframe with pd.json_normalize,
    # also making sure that there is a link to the main table.
prod = pd.json_normalize(data = data, record_path = "production_companies", meta = "id", record_prefix = "comp_")
prod
```

```python
# Now movies data frame has been splitted into a 4 data frames.
# Now also starting each and every table in data base movies.db

# First of all creating connection with connect method

con = sq3.connect("movies.db")
```

```python
con
```

```python
# Loading data frame movies into database
    # Inside brackets defining table name, passing connection object, and drop a range index.
movies.to_sql("Movies", con, index = False)
```

```python
votes.to_sql("Votes", con, index = False)
```

```python
genres.to_sql("Genres", con, index = False)
```

```python
prod.to_sql("Prod", con, index = False)
```

```python
con.execute("Select * FROM sqlite_master").fetchall()
```

```python
# Getting names of the tables from the database.
con.execute("SELECT name FROM sqlite_master WHERE type='table' ORDER BY name").fetchall()
```

```python
con.close()
```

### Unconditional filtering of database

Thus, importing only those rows or movies sent to Python and Pandas that meet certain conditions.

If there are very large data sets, most of the data is kept in the database, an donly load those data into python and pandas that we are interested in.

This is not only memory efficient, but also imporoves the performance of Python and Pandas.

```python
con = sq3.connect("movies.db")
```

```python
# Filtering the movies table only for those rows with movie id  597
    # 3 key words
        # SELECT
        # FROM Movies table
        # WHERE id = 597
# Python and Pandas can handle lowercase letters.
pd.read_sql("SELECT * FROM Movies WHERE id = 597", con)
```

```python
# Selecting all movies and all columns where there is:
    # revenue greater than 2000
pd.read_sql("SELECT * FROM Movies WHERE revenue > 2000", con)
```

```python
# Selecting all movies with revenue greater than 1500 OR budget < than 200 million
pd.read_sql("SELECT * FROM Movies WHERE revenue > 1500 OR budget < 200", con)
```

```python
# Select movies where the revenue is greater than 1250, minimum valu ein the budged column
pd.read_sql("SELECT MIN(budget) FROM Movies WHERE revenue > 1250", con)
```

```python
# Unique values in a specific column with the key word DISTINCT
# With SELECT DISTINCT title, we get unique movie titles in the column title from the movies table.
pd.read_sql("SELECT DISTINCT title FROM Movies", con)
```

```python
# Selecting unique values in belongs_to_collection_name
pd.read_sql("SELECT DISTINCT belongs_to_collection_name FROM Movies", con)
```

```python
# Selecting all columns, rows from the movies table, 
        # and ordering the movie rows by the budget in descending order.
    # 2 key words ORDER BY and DESC
pd.read_sql("SELECT * FROM Movies ORDER BY budget DESC", con)
```

```python
# SELECT those moviess, rows with missing values in column belongs_to_collection_name
    #(those moies that are actually standalone movies)
pd.read_sql("SELECT * FROM Movies WHERE belongs_to_collection_name IS NULL", con)
```

```python
# Also the opposite, all rows, all movies where there is no missing values
    # in belongs_to_collection_column
pd.read_sql("SELECT * FROM Movies WHERE belongs_to_collection_name IS NOT NULL", con)
```

```python
# All movies that have the following pattern in the title 'Avengers%'
    # % - percentage sign is the wild card for one or many characters.
pd.read_sql("SELECT * FROM Movies WHERE title LIKE 'Avengers%'", con)
```

```python
con.close()
```

### Join Queries

```python
con = sq3.connect("movies.db")
```

```python
# Movies table
pd.read_sql("SELECT * FROM Movies", con)
```

```python
# Votes table
pd.read_sql("SELECT * FROM Votes", con)
```

```python
# In this simple case do not have to differentiate between inner joints or left joints.
# Therefore Merging both dataframes on the movie ID.
    # SELECT * - selecting all
    # FROM Movies - from where
    # JOIN Votes - Joining Movies to Votes
    # ON - joining on specific column (ID column in the movies table equals ID column in the votes table)

pd.read_sql("SELECT * \
            FROM Movies \
            JOIN Votes \
            ON Movies.id=Votes.id", con)
```

```python
# Selecting only the ID column in the movies table, title and vote average
pd.read_sql("SELECT Movies.id, Movies.title, Votes.vote_average \
            FROM Movies \
            JOIN Votes \
            ON Movies.id=Votes.id", con, index_col = "id")
```

```python
# Conditional filtering
# With movies with vote average of greater than 8
pd.read_sql("SELECT Movies.id, Movies.title, Votes.vote_average \
           FROM Movies \
           JOIN Votes \
           ON Movies.id=Votes.id \
           WHERE Votes.vote_average > 8", con, index_col = "id")
```

```python
# Sort the movies by the budget in ascending order from low to high budged
pd.read_sql("SELECT Movies.id, Movies.title, Movies.budget, Votes.vote_average \
            FROM Movies \
            JOIN Votes \
            ON Movies.id=Votes.id \
            WHERE Votes.vote_average > 8 \
            ORDER BY Movies.budget ASC", con, index_col = "id")
```

```python
con.close()
```

### Advanced Data Analysis

What are the most successful production companies in terms of total revenue.

The plan is to join the tables: production companies and movies and load join table into pandas.

Then group and agregate the data with pandas.

```python
con = sq3.connect("movies.db")
```

```python
# Production companies for each and every movie
pd.read_sql("SELECT * FROM Prod", con)
```

```python
# Want to get movie id, company name, movie revenue and movie title and save it in df.
# with inner join which is by default join works as well as this is not a complex daa.
df = pd.read_sql("SELECT Prod.id, Prod.comp_name, Movies.revenue, Movies.title \
            FROM Prod \
            LEFT JOIN Movies \
            ON Prod.id=Movies.id", con)
df
```

```python
# Grouping df by production companies and calculate the sum of revenue, and sort the results from high to low.
df.groupby("comp_name").revenue.sum().sort_values(ascending = False)
```

```python
# Assuming that we are interested in the production companies that were involved in the movie Titanic.
pd.read_sql("SELECT  Prod.comp_name \
            FROM Prod \
            LEFT JOIN Movies \
            ON Prod.id=Movies.id \
            WHERE Movies.title = 'Titanic'", con)
```

```python
# Assuming an interest with most successful genres
# All combination of genres and movies ids
df2 = pd.read_sql("SELECT Genres.id, Genres.genre_name, Movies.revenue, Movies.title \
            FROM Genres \
            LEFT JOIN Movies \
            ON Genres.id=Movies.id", con)
df2
```

```python
# Grouping df2 by genre name and agregated revenue, and get those with highest a the top.
df2.groupby("genre_name").revenue.sum().sort_values(ascending = False)
```

```python
# Genres for specific movie like Frozen 2, so only genre name swith movie title Frozen 2
pd.read_sql("SELECT Genres.genre_name \
            FROM Genres \
            LEFT JOIN Movies \
            ON Genres.id=Movies.id \
            WHERE Movies.title = 'Frozen II'", con)
```
